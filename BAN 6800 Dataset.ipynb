```python
df = pd.read_csv('customer_churn_dataset.csv')
```

```python
df.head()
```

```python
# Convert TotalCharges to numeric (if necessary)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')


```

```python
# Check for nulls and handle them
print("Null values per column:")
print(df.isnull().sum())

df.dropna(inplace=True)  # Drop rows with missing values
```

```python
# Exploratory Data Analysis
# Churn distribution
sns.countplot(x='Churn', data=df)
plt.title('Churn Distribution')
plt.show()
```

```python
# Tenure vs. MonthlyCharges by Churn
sns.scatterplot(x='Tenure', y='MonthlyCharges', hue='Churn', data=df)
plt.title('Tenure vs MonthlyCharges by Churn')
plt.show()
```

```python
# Support Tickets distribution
sns.boxplot(x='Churn', y='SupportTickets', data=df)
plt.title('Support Tickets by Churn')
plt.show()
```
